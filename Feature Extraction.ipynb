```python
import cv2
from skimage.feature import local_binary_pattern
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.filters import prewitt_h,prewitt_v
%matplotlib inline
```

```python
def euclideanDistance(nilai):
    return np.linalg.norm(nilai)
```

```python
numPoints = 3*8
radius = 3
```

```python
kotak = cv2.imread('kotak.jpg')
lingkaran = cv2.imread('lingkaran.jpg')
segitiga = cv2.imread('segitiga.jpg')
```

```python
plt.imshow(kotak)
```

```python
plt.imshow(lingkaran)
```

```python
plt.imshow(segitiga)
```

```python
kotak.shape
```

```python
lingkaran.shape
```

```python
segitiga.shape
```

# Convert to 2D array

```python
kotakGray = cv2.cvtColor(kotak, cv2.COLOR_RGB2GRAY)
lingkaranGray = cv2.cvtColor(lingkaran, cv2.COLOR_RGB2GRAY)
segitigaGray = cv2.cvtColor(segitiga, cv2.COLOR_RGB2GRAY)
```

```python
kotakGray
```

```python
lingkaranGray
```

```python
segitigaGray
```

## LBP

```python
lbpKotak = local_binary_pattern(kotakGray, numPoints, radius, method="uniform")
```

```python
lbpKotak
```

```python
plt.imshow(lbpKotak)
```

```python
lbpLingkaran = local_binary_pattern(lingkaranGray, numPoints, radius, method="uniform")
```

```python
lbpLingkaran
```

```python
plt.imshow(lbpLingkaran)
```

```python
lbpSegitiga = local_binary_pattern(segitigaGray, numPoints, radius, method="uniform")
```

```python
lbpSegitiga
```

```python
plt.imshow(lbpSegitiga)
```

## COLOR

```python
colorKotak = np.zeros((490,630)) 
colorLingkaran = np.zeros((374,750)) 
colorSegitiga = np.zeros((393,700)) 
```

```python
for i in range(0,kotak.shape[0]):
    for j in range(0,kotak.shape[1]):
        colorKotak[i][j] = ((kotak[i,j,0] + kotak[i,j,1] + kotak[i,j,2])/3)

colorKotak
```

```python
for i in range(0,lingkaran.shape[0]):
    for j in range(0,lingkaran.shape[1]):
        colorLingkaran[i][j] = ((lingkaran[i,j,0] + lingkaran[i,j,1] + lingkaran[i,j,2])/3)

colorLingkaran
```

```python
for i in range(0,segitiga.shape[0]):
    for j in range(0,segitiga.shape[1]):
        colorSegitiga[i][j] = ((segitiga[i,j,0] + segitiga[i,j,1] + segitiga[i,j,2])/3)
        
colorSegitiga
```

```python
plt.imshow(colorKotak)
```

```python
plt.imshow(colorLingkaran)
```

```python
plt.imshow(colorSegitiga)
```

## SHAPE

```python
shapeKotakH = prewitt_h(kotakGray)
shapeKotakV = prewitt_v(kotakGray)
shapeKotakH
```

```python
imshow(shapeKotakH, cmap='gray')
```

```python
shapeLingkaranH = prewitt_h(lingkaranGray)
shapeLingkaranV = prewitt_v(lingkaranGray)
shapeLingkaranH
```

```python
imshow(shapeLingkaranH, cmap='gray')
```

```python
shapeSegitigaH = prewitt_h(segitigaGray)
shapeeSegitigaV = prewitt_v(segitigaGray)
shapeSegitigaH
```

```python
imshow(shapeSegitigaH, cmap='gray')
```

```python
lbpKotak = np.reshape(lbpKotak, (490*630))
lbpLingkaran = np.reshape(lbpLingkaran, (374*750))
lbpSegitiga = np.reshape(lbpSegitiga, (393*700))
```

```python
colorKotak = np.reshape(colorKotak, (490*630)) 
colorLingkaran = np.reshape(colorLingkaran, (374*750))
colorSegitiga = np.reshape(colorSegitiga, (393*700))
```

```python
shapeKotak = np.reshape(shapeKotakH, (490*630))
shapeLingkaran = np.reshape(shapeLingkaranH, (374*750))
shapeSegitiga = np.reshape(shapeSegitigaH, (393*700))
```

```python
(uniqLKotak, countLKotak) = np.unique(lbpKotak, return_counts=True)
(uniqCKotak, countCKotak) = np.unique(colorKotak, return_counts=True)
(uniqSKotak, countSKotak) = np.unique(shapeKotak, return_counts=True)
```

```python
(uniqLLingkaran, countLLingkaran) = np.unique(lbpLingkaran, return_counts=True)
(uniqCLingkaran, countCLingkaran) = np.unique(colorLingkaran, return_counts=True)
(uniqSLingkaran, countSLingkaran) = np.unique(shapeLingkaran, return_counts=True)
```

```python
(uniqLSegitiga, countLSegitiga) = np.unique(lbpSegitiga, return_counts=True)
(uniqCSegitiga, countCSegitiga) = np.unique(colorSegitiga, return_counts=True)
(uniqSSegitiga, countSSegitiga) = np.unique(shapeSegitiga, return_counts=True)
```

```python
fig,ax = plt.subplots(1,3)
ax[0].plot(uniqLKotak,countLKotak)
ax[0].set_title("LBP KOTAK")
ax[1].plot(uniqCKotak,countCKotak)
ax[1].set_title("COLOR KOTAK")
ax[2].plot(uniqSKotak,countSKotak)
ax[2].set_title("SHAPE KOTAK")
plt.savefig('plot-fitur-kotak.png')
plt.show()
```

```python
fig,ax2 = plt.subplots(1,3)
ax2[0].plot(uniqLLingkaran,countLLingkaran)
ax2[0].set_title("LBP LINGKARAN")
ax2[1].plot(uniqCLingkaran,countCLingkaran)
ax2[1].set_title("COLOR LINGKARAN")
ax2[2].plot(uniqSLingkaran,countSLingkaran)
ax2[2].set_title("SHAPE LINGKARAN")
plt.savefig('plot-fitur-lingkaran.png')
plt.show()
```

```python
fig,ax3 = plt.subplots(1,3)
ax3[0].plot(uniqLSegitiga,countLSegitiga)
ax3[0].set_title("LBP SEGITIGA")
ax3[1].plot(uniqCSegitiga,countCSegitiga)
ax3[1].set_title("COLOR SEGITIGA")
ax3[2].plot(uniqSSegitiga,countSSegitiga)
ax3[2].set_title("SHAPE SEGITIGA")
plt.savefig('plot-fitur-segitiga.png')
plt.show()
```

```python
print("Berikut nilai euclidean untuk setiap feature extraction citra 1 (Buku | Kotak) : ")
print("LBP :",euclideanDistance(lbpKotak))
print("COLOR :",euclideanDistance(colorKotak))
print("SHAPE :",euclideanDistance(shapeKotak))
```
